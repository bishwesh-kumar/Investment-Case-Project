```python
# Supress Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import the numpy and pandas packages
import numpy as np
import pandas as pd
```

```python
#Import rounds2 csv data into dataframe
rounds2= pd.read_csv('rounds2.csv',encoding = "latin-1")
rounds2.head()
```

```python
#Rename column of rounds2 df
rounds2 = rounds2.rename(columns={'company_permalink': 'permalink'})
rounds2.head(2)
```

```python
#Encoding permalink and making uniform across the values(to lower case) in the df
rounds2['permalink'] = rounds2.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
rounds2['permalink']= rounds2['permalink'].str.lower()
```

```python
#Inspection of df
print(rounds2.shape)
print(rounds2.describe())
print(rounds2.columns)
print(rounds2.info)
```

```python
#loading companies csv into Dataframe
companies=pd.read_csv('companies.txt',sep="\t",encoding = "latin-1")
companies.head()
```

```python
#Inspecting companies df
print(companies.shape)
print(companies.describe())
print(companies.columns)
```

```python
#Encoding companies df- permalink and making uniform across the values (to lower case) in the df
companies['permalink']= companies['permalink'].str.lower()
companies['permalink'] = companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
```

```python
#Finding unique values in companies df
companies['name'].str.lower().nunique()
```

```python
#Finding unique values in rounds2 df
rounds2['permalink'].str.lower().nunique()
```

```python
#finding if there are any records in rounds2 which are not present in companies
master_f= pd.merge(rounds2,companies,how='outer',on ='permalink')
master_f.ix[master_f['permalink'].isnull(),:]
```

```python
#Merging both the df's on inner join & on the column permalink
master_frame= pd.merge(rounds2,companies,how='inner',on ='permalink')
master_frame.head()
```

```python
print(master_frame.shape)
```

```python
# To be on safe side, again encoding master_frame 
master_frame['permalink'] = master_frame.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
```

```python
#Verifying null counts in the df, column wise
master_frame.isnull().sum()
```

```python
#Checking for all NAN values percentages
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#removing rows
master_frame=master_frame[master_frame.isnull().sum(axis=1) <= 5]
```

```python
#Checking after removing the rows
master_frame.shape
```

```python
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#Removing null value records for raised_amount_usd as this column values cannot be imputed with mean/max/ min values. Hence drop
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
master_frame.shape
#master_frame.head()
```

```python
#Dropping other columns which have null values & doesnot require for our analysis
drop_columns=['funding_round_code','status','state_code','founded_at','funded_at','region','homepage_url','city']
master_frame=master_frame.drop(drop_columns,axis=1)
```

```python
#Df inspection after performing the drop operation
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
master_frame.shape
```

```python
#Converting into millions for easy calculation
master_frame.raised_amount_usd=(master_frame.raised_amount_usd/1000000)
```

```python
#Group by funding_round_type to know the funding type
df_grp=master_frame.groupby('funding_round_type')
```

```python
# to find the average values to drill down the funding type
amount_mean=df_grp.raised_amount_usd.mean().sort_values(ascending=False)
amount_mean
```

```python
#Now we know 'venture' is the funding type, lets create seperate df for this funding type & for further purpose
df1=master_frame[master_frame["funding_round_type"]=="venture"]
df1.shape
```

```python
#Grouping by countries to know the top 3 countries
df2=df1.groupby('country_code')
```

```python
# Creating a new df 'top9' to store the required top9 countries
top9=pd.DataFrame(df2.raised_amount_usd.sum().sort_values(ascending=False)).head(9)
top9
```

```python
#To split the sector & store it in new df, df3
df3=master_frame["category_list"].str.split("|",n=1, expand = True).rename(columns=lambda x: f"category_{x+1}")
master_frame['primary_sector']=df3.category_1
master_frame.head()
```

```python
#Loading the mappings sheet & renaming the columns accoridngly as per out usage
major_sector=pd.read_csv('mapping.csv')
major_sector
major_sector = major_sector.rename(columns={'category_list': 'primary_sector'})
major_sector.head()
```

```python
#Using melt function for major_sector df, to interchange rows to columns
major_sector
major_sector.set_index('primary_sector')
major_sector_melt= pd.melt(major_sector,id_vars=['primary_sector'],var_name='metrics', value_name='values')
```

```python
# As part of melt, we did recieve few unnecessary values, deleting them
major_sector_melt==0
major_sector_melt = major_sector_melt[~(major_sector_melt == 0).any(axis=1)]
major_sector_melt.shape
```

```python
#Merging master frame & major_sector df's. Structuring the master_frame2 correctly for further usage & for readability
master_frame2=pd.merge(master_frame,major_sector_melt,how="inner",on="primary_sector")
master_frame2
master_frame2 = master_frame2.drop('values', axis=1)
master_frame2 = master_frame2.rename(columns={'metrics': 'main_sector'})
master_frame2.shape
```

```python
#Inspecting new df
master_frame2.isnull().sum()
```

Now we have to create 3 Data frames for the countries : USA, GBR & IND

```python
# Data frame for USA
D1= master_frame2.loc [(master_frame2.funding_round_type=="venture") & (master_frame2.country_code=="USA") & (master_frame2.raised_amount_usd >5) & (master_frame2.raised_amount_usd <15)]
D1.shape
```

```python
# Data frame for GBR
D2= master_frame2.loc [(master_frame2.funding_round_type=="venture") & (master_frame2.country_code=="GBR") & (master_frame2.raised_amount_usd >5) & (master_frame2.raised_amount_usd <15)]
D2.shape
```

```python
# Data frame for IND
D3= master_frame2.loc [(master_frame2.funding_round_type=="venture") & (master_frame2.country_code=="IND") & (master_frame2.raised_amount_usd >5) & (master_frame2.raised_amount_usd <15)]
D3.shape
```

```python
#Grouping D1 dataframe by 'main_sector' to know the count of investments & sum 
D1_investements=D1.groupby('main_sector').main_sector.count()
D1_Amount=D1.groupby('main_sector').raised_amount_usd.sum()

#Transform is used to include the values in D1 recieved above 
D1['Investements']=D1.groupby('main_sector').main_sector.transform('count')
D1['Amount']=D1.groupby('main_sector').raised_amount_usd.transform('sum')

print(D1_investements)
print(D1_Amount)

#To know which company recieved highest investement
D1[D1.main_sector=="Others"].sort_values(by="raised_amount_usd",ascending= False)
D1[D1.main_sector=="Cleantech / Semiconductors"].sort_values(by="raised_amount_usd",ascending= False)
```

```python
#Grouping D2 dataframe by 'main_sector' to know the count of investments & sum 
D2_investements=D2.groupby('main_sector').main_sector.count()
D2_Amount=D2.groupby('main_sector').raised_amount_usd.sum()

#Transform is used to include the values in D1 recieved above 
D2['Investements']=D2.groupby('main_sector').main_sector.transform('count')
D2['Amount']=D2.groupby('main_sector').raised_amount_usd.transform('sum')

print(D2_investements)
print(D2_Amount)

#To know which company recieved highest investement
D2[D2.main_sector=="Others"].sort_values(by="raised_amount_usd",ascending= False)
D2[D2.main_sector=="Cleantech / Semiconductors"].sort_values(by="raised_amount_usd",ascending= False)
```

```python
#Grouping D3 dataframe by 'main_sector' to know the count of investments & sum 
D3_investements=D3.groupby('main_sector').main_sector.count()
D3_Amount=D3.groupby('main_sector').raised_amount_usd.sum()

#Transform is used to include the values in D1 recieved above 
D3['Investements']=D3.groupby('main_sector').main_sector.transform('count')
D3['Amount']=D3.groupby('main_sector').raised_amount_usd.transform('sum')

print(D3_investements)
print(D3_Amount)

#To know which company recieved highest investement
D3[D3.main_sector=="Others"].sort_values(by="raised_amount_usd",ascending= False)
D3[D3.main_sector=="News, Search and Messaging"].sort_values(by="raised_amount_usd",ascending= False)
```
